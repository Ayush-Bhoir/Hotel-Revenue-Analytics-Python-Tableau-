```python
dup = df.duplicated().sum()
print(f"Duplicate counts are : {dup}")
```

```python
df.drop_duplicates(inplace = True)
```

```python
# In my dataset there were around 1.20 lakh rows , When i remove the duplicate rows i left with around 87k rows
df.shape
```

```python
df.isna().sum()
```

```python
#   children - 4            -------  int
#   country - 452           -------  obj
#   agent - 12193           -------  int
#   company - 82137
```

```python
# There was 4 missing values in children and it is numeric column , So we can take mean()
df['children'] = df['children'].fillna(df['children'].mean())

# 452 Null values in Country and country cannot be Null , I replace it with "Other"
df['country'] = df['country'].fillna('Other')

# 12193 Null values in Agent column
# agent is in numeric column so we can take mean() of column to fill Null values
df['agent'] = df['agent'].fillna(df['agent'].mean())

# Out of 87396 rows in dataset .  [company] has 82137 null values , So it has maximum null values 
# Thats why i removed this column
df = df.drop('company' , axis = 1)
```

```python
df.shape
```

```python
df.isna().sum()
```

```python
df.dtypes
```

```python
# Changing the [0 , 1] values with ["Yes" , "No"]

df['is_canceled'] =df['is_canceled'].replace([0,1] , ["No","Yes"])
df['is_repeated_guest'] = df['is_repeated_guest'].replace([0,1] , ["No","Yes"])
```

```python
df['children'] = df['children'].astype(int)
df['agent'] = df['agent'].astype(int)
df['reservation_status_date'] = pd.to_datetime(df['reservation_status_date'] , dayfirst=True)
df['is_canceled'] = df['is_canceled'].astype(object)
df['is_repeated_guest'] = df['is_repeated_guest'].astype(object)
```

# Visuals

```python
# 1) What proportion of bookings were canceled versus not canceled?

percentage = df['is_canceled'].value_counts()
plt.pie(percentage ,
        colors=['mediumspringgreen','skyblue'] ,
        autopct = '%1.1f%%' ,
        labels = percentage.index ,
        wedgeprops={'width': 0.5}
        )
plt.title('Hotel Type Distribution')
plt.show()
```

```python
# 2) How do cancellation rates differ between hotel types (City vs. Resort)?

plt.figure(figsize=(10,6))
hotel_cancel_rate = df.groupby(['hotel', 'is_canceled']).size().reset_index(name='count')
hotel_total = df.groupby('hotel').size().reset_index(name='total')
hotel_cancel_rate = hotel_cancel_rate.merge(hotel_total, on='hotel')
hotel_cancel_rate['proportion'] = hotel_cancel_rate['count'] / hotel_cancel_rate['total']

a = sns.barplot(data=hotel_cancel_rate, x='hotel', y='proportion', hue='is_canceled', palette='pastel')
plt.title("Cancellation Rate by Hotel Type")
plt.ylabel("Proportion")
plt.xlabel("Hotel Type")

for bars in a.containers:
    a.bar_label(bars, labels=[f'{v.get_height() * 100:.1f}%' for v in bars], padding=3)

plt.show()
```

```python
# 3) How does the Average Daily Rate (ADR) trend over time for City hotels compared to Resort hotels?

plt.figure(figsize = (10,6) )
df['Month_Year'] = df['reservation_status_date'].dt.to_period('M').astype(str)
monthly_adr = df.groupby(['Month_Year', 'hotel'])['adr'].mean().reset_index()

sns.lineplot(data=monthly_adr, x='Month_Year', y='adr', hue='hotel', marker="o")
plt.title("ADR Trend Over Time by Hotel Type")
plt.ylabel("Average Daily Rate (ADR)")
plt.xlabel("Month-Year")
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()

```

```python
# 4) How do reservation and cancellation counts vary month‑by‑month?
plt.figure(figsize = (10,6) )

df['Month'] = df['reservation_status_date'].dt.strftime('%b')


month_cancellation = df.groupby(['Month', 'is_canceled']).size().reset_index(name='count')

a = sns.barplot(data=month_cancellation, x='Month', y='count', hue='is_canceled', palette='coolwarm')
plt.title("Monthly Reservation and Cancellation Counts")
plt.xlabel("Month")
plt.ylabel("Count")

for bars in a.containers :
    a.bar_label(bars)

plt.grid(True)
plt.show()


```

```python
# 5) Which top 10 countries have the most canceled reservations?

plt.figure(figsize = (10,6) )

top_countries = df[df['is_canceled'] == 'Yes']['country'].value_counts().head(10).reset_index()
top_countries.columns = ['Country', 'Cancellations']

a = sns.barplot(data=top_countries, y='Country', x='Cancellations', palette='magma')
plt.title("Top 10 Countries with Most Canceled Reservations")
plt.xlabel("Number of Cancellations")
plt.ylabel("Country")

for bars in a.containers :
    a.bar_label(bars)

plt.show()

```
